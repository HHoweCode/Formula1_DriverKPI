```python
# Filter for the last 5 years using df_race
recent_years = df_race[df_race['year'] >= df_race['year'].max() - 4]
recent_race_ids = recent_years['raceid'].unique()

# Filter results for recent years
df_result_recent = df_result[df_result['raceid'].isin(recent_race_ids)]

# Merge race results with race, driver, and constructor info
df_merged = df_result_recent.merge(df_race[['raceid', 'year']], on='raceid')\
                            .merge(df_driver, on='driverid')\
                            .merge(df_constructor, on='constructorid')

# ----- DRIVER KPIs -----
driver_kpis = df_merged.groupby(['driverid', 'surname', 'year']).agg(
    total_points=('points', 'sum'),
    total_wins=('positionorder', lambda x: (x == 1).sum()),
    podiums=('positionorder', lambda x: (x <= 3).sum()),
    avg_finish=('positionorder', 'mean'),
    dnfs=('statusid', lambda x: (x != 1).sum())  # Assuming statusId 1 = Finished
).reset_index()

# ----- CONSTRUCTOR KPIs -----
constructor_kpis = df_merged.groupby(['constructorid', 'name', 'year']).agg(
    total_points=('points', 'sum'),
    total_wins=('positionorder', lambda x: (x == 1).sum()),
    podiums=('positionorder', lambda x: (x <= 3).sum()),
    avg_finish=('positionorder', 'mean'),
    dnfs=('statusid', lambda x: (x != 1).sum())
).reset_index()
```

```python
# Merge lap times with driver and race info
df_laps = df_time.merge(df_driver, on='driverid')\
                 .merge(df_race[['raceid', 'year']], on='raceid')

# Optional: convert time columns if needed (time is in milliseconds already, no conversion needed)

# Group by driver and race to calculate lap time consistency
lap_consistency = df_laps.groupby(['driverid', 'surname', 'raceid', 'year']).agg(
    avg_lap_time=('time', 'mean'),
    std_lap_time=('time', 'std'),
    laps_completed=('lap', 'count')
).reset_index()

# Filter out races where driver completed very few laps (optional threshold, say < 5)
lap_consistency = lap_consistency[lap_consistency['laps_completed'] >= 5]
```

```python
lap_consistency.head()
```

```python
season_lap_consistency = lap_consistency.groupby(['driverid', 'surname', 'year']).agg(
    avg_std_lap_time=('std_lap_time', 'mean'),
    total_races=('raceid', 'nunique')
).reset_index()
```

```python
# Group by year to visualize top performers
top10_points = driver_kpis.sort_values(by='total_points', ascending=False).head(10)

# Plot: Total Points for Top 10 Drivers
plt.figure(figsize=(10, 6))
plt.bar(top10_points['surname'], top10_points['total_points'])
plt.title('Top 10 Drivers by Total Points (2021)')
plt.xlabel('surname')
plt.ylabel('total_points')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot: Podiums for Top 10 Drivers
plt.figure(figsize=(10, 6))
plt.bar(top10_points['surname'], top10_points['podiums'])
plt.title('Top 10 Drivers by Podium Finishes (2021)')
plt.xlabel('surname')
plt.ylabel('podiums')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot: Average Finish Position (lower is better)
plt.figure(figsize=(10, 6))
plt.bar(top10_points['surname'], top10_points['avg_finish'])
plt.title('Top 10 Drivers by Average Finish Position (2021)')
plt.xlabel('surname')
plt.ylabel('avg_finish')
plt.xticks(rotation=45)
plt.gca().invert_yaxis()  # lower is better
plt.tight_layout()
plt.show()
```

```python
# Line Plot: Total Points Over Years
plt.figure(figsize=(10, 6))
for driver in driver_kpis['surname'].unique():
    df = driver_kpis[driver_kpis['surname'] == driver]
    plt.plot(df['year'], df['total_points'], marker='o', label=driver)

plt.title('Total Points Over the Years')
plt.xlabel('year')
plt.ylabel('total_points')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Line Plot: Total Wins Over Years
plt.figure(figsize=(10, 6))
for driver in driver_kpis['surname'].unique():
    df = driver_kpis[driver_kpis['surname'] == driver]
    plt.plot(df['year'], df['total_wins'], marker='o', label=driver)

plt.title('Total Wins Over the Years')
plt.xlabel('year')
plt.ylabel('total_wins')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
